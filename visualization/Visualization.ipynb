```python
import json

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
#import matplotlib
import seaborn as sns

import visual_funcs as vis
```

```python
file = open(r'../data/planesnet.json')
planesnet = json.load(file)
file.close()
print(planesnet.keys())
```

```python
df = pd.DataFrame(planesnet)
df.head()
```

```python
print("Total number of planes images: ", df.shape[0])
ac_count = df['labels'].value_counts()
plt.figure(figsize=(12,8))
sns.barplot(x=ac_count.index, y=ac_count.values)
plt.title("Image Count (1: Plane. 0: Not Plane)", fontsize=16)
plt.xlabel("Label", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.show()
```
